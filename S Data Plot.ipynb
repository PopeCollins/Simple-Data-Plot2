### THIS IS A SIMPLE PLOT OF DATA

 This shows how data can be plot using various descriptive analysis methods

```python
## Display multiple plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.array([0,2,3,4])
y =np.array([3,6,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

x1 = np.array([0,1,2,3])
y1 = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x1,y1)

x3= np.array([0,1,2,3])
y3 = np.array([30,45,55,70])
plt.subplot(1,2,2)
plt.plot(x3,y3)
plt.show()
```

```python
## Scatter plot
```

```python
x = np.array([2,3,4,5,6,1,4,7,6,2,4,6,8,7])
y = np.array([23,45,66,12,67,67,88,90,22,445,56,34,45,67])
plt.subplot(1,2,1)
plt.scatter(x,y, c="g", s=60)       #c=r,b  (color) s=size
```

```python
## Bar plot
```

```python
x = np.array(["Shoe", "Cloth","Phone","Bag"])
y = np.array([23,29,30,16])
plt.bar(x,y, color="purple")

```

```python
# for Horrizontal bar

x1 = np.array(["Shoe", "Cloth","Phone","Bag"])
y1 = np.array([23,29,30,16])
plt.barh(x1,y1, color="orange")
```

```python
# for pie chart

namepie=["Food","Juice","Wine","Alcohol"]
plt.pie(y, labels=namepie)
plt.show()
```
